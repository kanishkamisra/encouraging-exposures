```python
import csv
import pathlib
import re
import utils

from collections import defaultdict, Counter
# from minicons.utils import find_pattern
from spacy.lang.en import English
from tqdm import tqdm
import spacy
```

```python
nlp = English()
tokenizer = nlp.tokenizer

def tokenize(string):
    return [t.text for t in tokenizer(string)]

def whitespace_tokenize(string):
    return [t.text + t.whitespace_ for t in tokenizer(string)]
```

```python
aochildes = utils.read_file("../data/corpora/aochildes.train")
postags = utils.read_file("../data/corpora/aochildes.postags")
```

```python
sent_tokens = [whitespace_tokenize(sent) for sent in aochildes]
postag_tokens = [postag.split(" ") for postag in postags]
```

```python
postag_token_stats = defaultdict(lambda : defaultdict(int))

for sent, postagseq in zip(sent_tokens, postag_tokens):
    for word, postag in zip(sent, postagseq):
        postag_token_stats[postag][word] += 1

for k,v in postag_token_stats.items():
    v = dict(v)

postag_token_stats = dict(postag_token_stats)
```
