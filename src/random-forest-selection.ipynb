```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import utils
```

```python
results = utils.read_csv_dict("../data/results/final_results_250716_coded.csv")
```

```python
FEATURES=['theme_pronominality', 'theme_animacy', 'theme_definiteness', 'theme_givenness', 'recipient_pronominality', 'recipient_animacy', 'recipient_definiteness', 'recipient_givenness', 'length_diff']

def select_exp(results, exp = "do -> pp"):
    features = []
    scores = []
    for item in results:
        if item['exp'] == exp:
            feature_vec = []
            for feature in FEATURES:
                feature_vec.append(item[feature])
            feature_vec = [int(x) for x in feature_vec]
            features.append(feature_vec)

            scores.append(float(item['score']))

    return np.array(features), np.array(scores)
```

```python
do_pp = select_exp(results, exp="do -> pp")
```

```python
X_train, y_train = do_pp
```

```python
rf = RandomForestRegressor(n_estimators=1000, random_state=0)
rf.fit(X_train, y_train)
```

```python
print(f"Train R^2: {rf.score(X_train, y_train):.3f}")
```

```python
possible_values = np.unique(X_train, axis=0)
```

```python
predictions = rf.predict(possible_values)

# max(results, key=lambda x: x[2])
# max(predictions)
predictions.argmin()
```
